```python
# Import the neccassary libraries

from langchain_huggingface import ChatHuggingFace
from langchain_huggingface import HuggingFaceEndpoint
```

```python
llm = HuggingFaceEndpoint(
    # repo_id="Qwen/Qwen2.5-7B-Instruct",
    repo_id="deepseek-ai/DeepSeek-R1",
    task="text-generation",
    max_new_tokens=512,
    do_sample=False,
    repetition_penalty=1.03,
    huggingfacehub_api_token=HF_READ_TOKEN
)
```

```python
chat_model = ChatHuggingFace(llm=llm)
chat_model.invoke("Hello how are you?")
```

```python
# Will discuss more in RAG chapter

from langchain_core.prompts import PromptTemplate
from langchain_core.output_parsers import StrOutputParser
from langchain_core.runnables import RunnablePassthrough

template="""Question: {question}

give me an answer in detail manner and in step by step manner."""

prompt = PromptTemplate(template=template, input_variables=["question"])

llm_chain = (
    {"question": RunnablePassthrough()}
    | prompt
    | chat_model
    | StrOutputParser()
)

question="who is the first president of Cameroon?"

llm_chain.invoke({"question": question})
```
